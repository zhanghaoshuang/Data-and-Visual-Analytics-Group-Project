```python
import pandas as pd
import numpy as np
import scipy as sp
```

```python
no_business= pd.read_csv("no_business.csv")
no_review=pd.read_csv("no_review.csv")
```

```python
no_review["year"]=no_review["date"].apply(lambda x: x[0:4])
business_year_open=no_review.groupby("business_id").min().reset_index()[["business_id","year"]]
business_year_open["year"]= business_year_open["year"].apply(lambda x:int(x))
business_year_open
```

```python
business_year_open_post_2020= business_year_open [business_year_open["year"]>=2020]
business_year_open_pre_2020= business_year_open [business_year_open["year"]<2020]
```

```python
business_year_open_post_2020_zip=business_year_open_post_2020.merge(no_business, on="business_id")[["business_id","year","postal_code"]]
business_year_open_post_2020_zip
```

```python
post_2020_new_business=business_year_open_post_2020_zip.groupby(["postal_code"]).count().reset_index()[["postal_code","business_id"]]
post_2020_new_business=post_2020_new_business.rename(columns={"business_id":"post_2020"})
post_2020_new_business=post_2020_new_business.sort_values("postal_code")
post_2020_new_business["postal_code"]=post_2020_new_business["postal_code"].apply(lambda x:int(x))
post_2020_new_business
```

```python
business_year_open_pre_2020_zip=business_year_open_pre_2020.merge(no_business, on="business_id")[["business_id","year","postal_code"]]
business_year_open_post_2020_zip
```

```python
pre_2020_new_business=business_year_open_pre_2020_zip.groupby(["postal_code"]).count().reset_index()[["postal_code","business_id"]]
pre_2020_new_business=pre_2020_new_business.rename(columns={"business_id":"pre_2020"})
pre_2020_new_business=pre_2020_new_business.sort_values("postal_code")
pre_2020_new_business["postal_code"]=pre_2020_new_business["postal_code"].apply(lambda x:int(x))
pre_2020_new_business
```

```python
businesses_by_year=pre_2020_new_business.merge(post_2020_new_business,on = "postal_code",how="outer")
businesses_by_year=businesses_by_year.fillna(0)
businesses_by_year
```

```python
no_demand_score= pd.read_csv("no_demand_score_pre_2020.csv")[["business_id","postal_code","total_demand_score","missing_demand_score"]]
```

```python
no_demand_score_zip=no_demand_score.groupby("postal_code").sum().reset_index()[["postal_code","total_demand_score","missing_demand_score"]]
no_demand_score_zip["postal_code"]=no_demand_score_zip["postal_code"].apply(lambda x:int(x))
no_demand_score_zip
```

```python
businesses_by_year_demand=no_demand_score_zip.merge(businesses_by_year,on="postal_code",how="inner")
businesses_by_year_demand=businesses_by_year_demand.sort_values("post_2020",ascending=False).reset_index(drop=True)
businesses_by_year_demand
```

```python
new_business_rank= list(businesses_by_year_demand.index)
missing_demand_rank=np.argsort(-np.array(businesses_by_year_demand["missing_demand_score"]))
```

```python
sp.stats.kendalltau(new_business_rank, missing_demand_rank)
```
